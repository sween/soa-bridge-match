```python
from fhir.resources.careplan import CarePlan
from fhir.resources.servicerequest import ServiceRequest

subjects = {}
encounters = []
study = ResearchStudy(status="completed")
study.id = "H2Q-MC-LZZT"
for record in sv.itertuples():
    if record.USUBJID not in subjects:
        patient = Patient()
        patient.id = record.USUBJID
        subject = ResearchSubject(status="off-study", 
            individual = Reference(reference=f"Patient/{patient.id}") , 
            study = Reference(reference=f"ResearchStudy/{study.id}"))
        subject.id = record.USUBJID
        subjects[record.USUBJID] = subject
    subject = subjects[record.USUBJID]
    careplan = CarePlan(status="completed")
    careplan.basedOn = Reference(reference=f"PlanDefinition/{record.PLANID}")
    service_request = ServiceRequest(status="completed")
    service_request.basedOn = Reference(reference=f"CarePlan/{careplan.id}")
    encounter = Encounter(status="finished", class_fhir=Coding(code="IMP", system="http://hl7.org/fhir/v3/ActCode"))
    encounter.id = f"{record.USUBJID}-{record.VISITNUM}"
    encounter.subject = Reference(reference=f"Patient/{patient.id}")
    period = {}
    if record.SVSTDTC:
        period["start"] = record.SVSTDTC
    if record.SVENDTC:
        period["end"] = record.SVENDTC
    encounter.period = period
    encounters.append(encounter)


  
    

```
